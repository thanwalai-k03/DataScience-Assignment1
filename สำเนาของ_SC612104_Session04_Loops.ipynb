````python
# วิธีไม่ใช้ loop - เขียนซ้ำๆ
print(1)
print(2)
print(3)
print(4)
print(5)
````

## 2. `range()` — ตัวช่วยสร้างลำดับเลข

`range()` เป็นฟังก์ชันที่สร้าง "ลำดับของเลขจำนวนเต็ม" ไว้ใช้คู่กับ `for` loop เป็นหลัก มี 3 รูปแบบการเรียกใช้:

| รูปแบบ | ความหมาย | ตัวอย่าง | ค่าที่ได้ |
|---|---|---|---|
| `range(stop)` | เริ่มจาก 0 ถึง stop-1 | `range(5)` | `0,1,2,3,4` |
| `range(start, stop)` | เริ่มจาก start ถึง stop-1 | `range(2, 6)` | `2,3,4,5` |
| `range(start, stop, step)` | เพิ่มทีละ step | `range(0, 10, 2)` | `0,2,4,6,8` |

**⚠️ จุดสำคัญที่สุด:** `range()` ไม่รวมค่า `stop` เสมอ! `range(1, 6)` จะได้ 1,2,3,4,5 (ไม่มี 6)


````python
type(list(range(5)))
````

````python
# range(stop) - เริ่มจาก 0
print(list(range(5)))     # [0, 1, 2, 3, 4]

# range(start, stop)
print(list(range(2, 6)))  # [2, 3, 4, 5]

# range(start, stop, step)
print(list(range(0, 10, 2)))   # [0, 2, 4, 6, 8]

# step เป็นค่าลบได้ -> นับย้อนกลับ
print(list(range(10, 0, -1)))  # [10, 9, ..., 1]
````

**หมายเหตุ:** ปกติเราไม่ต้องครอบ `range()` ด้วย `list()` — ที่ทำตรงนี้เพื่อให้เห็นค่าทั้งหมดที่ `range()` สร้างขึ้นเท่านั้น ในการใช้งานจริงกับ `for` loop ไม่ต้องใส่ `list()`

## 3. `for` Loop

ใช้ `for` loop เมื่อ**รู้จำนวนรอบที่แน่นอน** หรือต้องการวนข้อมูลในลำดับ (เช่น list, string, range)

### โครงสร้างพื้นฐาน

```python
for ตัวแปร in ลำดับข้อมูล:
    # ทำงานซ้ำสำหรับสมาชิกแต่ละตัวในลำดับ
```

เหมือนกับ `if`: ต้องมี `:` ท้ายบรรทัด และโค้ดด้านในต้องเยื้อง (indent)


````python
# for loop พื้นฐานกับ range()
for i in range(1, 6):
    print(f"รอบที่ {i}")
````

````python
# for loop วนผ่านตัวอักษรใน string
word = "Python"
for letter in word:
    print(letter)
````

````python
# for loop วนผ่านสมาชิกใน list
fruits = ["แอปเปิ้ล", "กล้วย", "ส้ม"]
for fruit in fruits:
    print(f"ผลไม้: {fruit}")
````

### for loop ซ้อนกัน (Nested for loop)

ใช้เมื่อต้องวนข้อมูล 2 มิติ เช่น ตารางคูณ, ตาราง 2D


````python
# ตารางสูตรคูณแม่ 2 ถึง 4 (ใช้ for ซ้อน for)
for table in range(2, 7, 2):           # loop รอบนอก: แม่สูตรคูณ
    print(f"--- {table} ---")
    for multiplier in range(1, 7):   # loop รอบใน: 1 ถึง 3
        result = table * multiplier
        print(f"{table} x {multiplier} = {result}")
````

## 4. `while` Loop

ใช้ `while` loop เมื่อ**ไม่รู้จำนวนรอบที่แน่นอน** แต่รู้ "เงื่อนไข" ที่จะให้ loop ทำงานต่อไป — loop จะทำงานซ้ำไปเรื่อยๆ **ตราบใดที่เงื่อนไขยังเป็นจริง**

### โครงสร้างพื้นฐาน

```python
while เงื่อนไข:
    # ทำงานซ้ำ ตราบใดที่เงื่อนไขเป็น True
```

**⚠️ จุดสำคัญที่สุด:** ต้องมีโค้ดที่ทำให้เงื่อนไขกลายเป็น `False` ในที่สุด ไม่อย่างนั้นจะเกิด **infinite loop** (วนซ้ำไม่มีที่สิ้นสุด) ค้างไปตลอด


````python
# while loop พื้นฐาน - นับ 1 ถึง 5
count = 1

while count <= 5:
    print(f"count = {count}")
    count += 1   # สำคัญมาก! ถ้าลืมบรรทัดนี้ จะเป็น infinite loop
````

### เปรียบเทียบ for กับ while

| | `for` loop | `while` loop |
|---|---|---|
| ใช้เมื่อ | รู้จำนวนรอบแน่นอน | ไม่รู้จำนวนรอบ รู้แต่เงื่อนไข |
| ตัวอย่างการใช้งาน | วนตามจำนวนข้อมูลใน list | รอ input จนกว่าจะถูกต้อง, เกมที่เล่นจนกว่าจะแพ้ |

ตัวอย่างด้านล่าง: จำลองสถานการณ์ "ถามรหัสผ่านจนกว่าจะถูก" — ไม่รู้ว่าผู้ใช้จะพิมพ์ถูกตอนไหน จึงไม่เหมาะกับ `for`


````python
# จำลอง while loop กับสถานการณ์ที่ไม่รู้จำนวนรอบล่วงหน้า
correct_password = "python2025"
attempts = ["12345", "admin", "python2025"]  # จำลองการพิมพ์ผิด 2 ครั้งก่อนถูก
i = 0

guess = attempts[i]
while guess != correct_password:
    print(f"รหัสผ่านผิด: {guess}")
    i += 1
    guess = attempts[i]

print(f"รหัสผ่านถูกต้อง: {guess} -- เข้าสู่ระบบสำเร็จ!")
````

## 5. `break` และ `continue`

ใช้ควบคุมการทำงานของ loop แบบพิเศษ — ทั้งสองคำสั่งใช้ได้ทั้งกับ `for` และ `while`

| คำสั่ง | ความหมาย |
|---|---|
| `break` | **หยุด loop ทันที** ออกจาก loop เลย ไม่ทำรอบที่เหลือ |
| `continue` | **ข้ามไปรอบถัดไป** ไม่ทำโค้ดที่เหลือในรอบนี้ แต่ loop ยังทำงานต่อ |


````python
# break: หยุด loop ทันทีที่เจอเงื่อนไข
print("ตัวอย่าง break:")
for num in range(1, 11):
    if num == 5:
        print("เจอ 5 แล้ว หยุด loop!")
        break
    print(num)

print("--- จบ loop ---")
````

````python
# continue: ข้ามรอบนี้ ไปทำรอบถัดไป
print("ตัวอย่าง continue: print เฉพาะเลขคี่")
for num in range(1, 11):
    if num % 2 == 0:   # ถ้าเป็นเลขคู่
        continue         # ข้ามไปรอบถัดไปทันที ไม่ print
    print(num)
````

### เปรียบเทียบ break vs continue ในภาพเดียว


````python
print("break ที่ num == 3:")
for num in range(1, 6):
    if num == 3:
        break
    print(num)
print()

print("continue ที่ num == 3:")
for num in range(1, 6):
    if num == 3:
        continue
    print(num)
````

**สังเกตผลต่าง:**
- `break` ที่ num==3 → print 1, 2 แล้วหยุดเลย (ไม่ทำรอบ 4, 5)
- `continue` ที่ num==3 → print 1, 2, 4, 5 (ข้ามแค่ 3 แต่ทำรอบ 4, 5 ต่อ)

### ตัวอย่างใช้งานจริง: ค้นหาในข้อมูล


````python
# ใช้ break เพื่อค้นหาและหยุดทันทีที่เจอ (มีประสิทธิภาพกว่าวนจนจบ)
students = ["มานี", "สมชาย", "วิชัย", "สมหญิง", "ปรีชา"]
target = "วิชัย"

for index, name in enumerate(students):
    if name == target:
        print(f"เจอ '{target}' ที่ตำแหน่ง index {index}")
        break
else:
    # else ของ for จะรันถ้า loop จบโดยไม่เจอ break เลย (ไม่บังคับต้องรู้ตอนนี้ก็ได้)
    print(f"ไม่เจอ '{target}' ในรายชื่อ")
````

> 💡 **เกร็ดเสริม:** `enumerate()` ที่เห็นในตัวอย่างข้างบน ใช้คู่กับ for loop เพื่อให้ได้ทั้ง index และค่า ไปพร้อมกันในรอบเดียว — มีประโยชน์มากเวลาอยากรู้ "ตำแหน่ง" ของข้อมูลด้วย ไม่ใช่แค่ค่า

## Comprehension

**Comprehension** คือวิธีเขียน loop แบบสั้นกระชับ เพื่อสร้าง list/dict/set ใหม่จากของเดิม — เขียนในบรรทัดเดียว แทนการเขียน `for` loop หลายบรรทัด


### List Comprehension

**โครงสร้างพื้นฐาน:**
```python
ผลลัพธ์ = [expression for item in iterable]
```

ลองดูปัญหา: อยากได้ list ของเลขยกกำลัง 2 จาก 1 ถึง 5

````python
# วิธีเดิม - ใช้ for loop ปกติ
squares = []
for num in range(1, 6):
    squares.append(num ** 2)
print("วิธีเดิม:", squares)

# วิธีใหม่ - list comprehension (บรรทัดเดียว!)
squares_v2 = [num ** 2 for num in range(1, 6)]
print("list comprehension:", squares_v2)
````

### List Comprehension พร้อมเงื่อนไข (filter)

**โครงสร้างที่มีเงื่อนไข:**
```python
ผลลัพธ์ = [expression for item in iterable if เงื่อนไข]
```


````python
# วิธีเดิม - กรองเอาเฉพาะเลขคู่ออกมาจาก 1-20
even_numbers = []
for num in range(1, 21):
    if num % 2 == 0:
        even_numbers.append(num)
print("วิธีเดิม:", even_numbers)

# วิธีใหม่ - list comprehension พร้อมเงื่อนไข
even_numbers_v2 = [num for num in range(1, 21) if num % 2 == 0]
print("list comprehension:", even_numbers_v2)
````

````python
# ตัวอย่างที่ใช้บ่อย: แปลงข้อมูลทั้ง list พร้อมกัน
names = ["manee", "somchai", "wichai"]

# แปลงเป็นตัวพิมพ์ใหญ่ทั้งหมด (ใช้กับ .upper() ของ string)
names_upper = [name.upper() for name in names]
print(names_upper)

# หาความยาวของแต่ละชื่อ
name_lengths = [len(name) for name in names]
print(name_lengths)

# กรองเฉพาะชื่อที่ยาวมากกว่า 4 ตัวอักษร
long_names = [name for name in names if len(name) > 4]
print(long_names)
````

### Dict Comprehension

**โครงสร้าง:**
```python
ผลลัพธ์ = {key_expr: value_expr for item in iterable}
```


````python
# สร้าง dict จาก list สองตัว (ชื่อ -> คะแนน)
names = ["มานี", "สมชาย", "วิชัย"]
scores = [85, 90, 78]

# วิธีเดิม - for loop
score_dict = {}
for i in range(len(names)):
    score_dict[names[i]] = scores[i]
print("วิธีเดิม:", score_dict)

# วิธีใหม่ - dict comprehension ใช้คู่กับ zip() (เรียนรายละเอียด zip ในหัวข้อถัดไป)
score_dict_v2 = {name: score for name, score in zip(names, scores)}
print("dict comprehension:", score_dict_v2)

# ตัวอย่าง: สร้าง dict เลขยกกำลัง 2 พร้อมกรองเงื่อนไข
squares_dict = {num: num ** 2 for num in range(1, 8) if num % 2 == 0}
print(squares_dict)   # {2: 4, 4: 16, 6: 36}
````

### Set Comprehension

**โครงสร้าง:**
```python
ผลลัพธ์ = {expression for item in iterable}
```

หน้าตาคล้าย dict comprehension มาก แต่ไม่มี `:` คั่น key-value (เพราะ set ไม่มี key-value)


````python
# หาความยาวที่ไม่ซ้ำกันของแต่ละชื่อ
names = ["มานี", "สมชาย", "วิชัย", "อร", "ปรีชา"]

unique_lengths = {len(name) for name in names}
print(unique_lengths)   # set ของความยาวที่ไม่ซ้ำกัน

# เทียบกับถ้าใช้ list comprehension (จะมีค่าซ้ำ)
all_lengths = [len(name) for name in names]
print(all_lengths)      # list ที่มีค่าซ้ำได้
````

### เมื่อไหร่ควร (และไม่ควร) ใช้ Comprehension

**ควรใช้เมื่อ:** logic ง่าย กระชับ อ่านแล้วเข้าใจได้ในแวบแรก

**ไม่ควรใช้เมื่อ:** logic ซับซ้อนมาก มี condition ซ้อนหลายชั้น — เขียนเป็น `for` loop ปกติจะอ่านง่ายกว่า

```python
# อ่านยาก เกินไป! ไม่ควรเขียนแบบนี้
result = [x**2 if x % 2 == 0 else x**3 for x in range(20) if x > 5 if x < 15]

# เขียนแบบนี้อ่านง่ายกว่ามาก
result = []
for x in range(20):
    if x > 5 and x < 15:
        if x % 2 == 0:
            result.append(x ** 2)
        else:
            result.append(x ** 3)
```

> 💡 **กฎง่าย ๆ:** ถ้า comprehension ของคุณยาวเกิน 1 บรรทัดปกติ หรือต้องอ่าน 2-3 รอบกว่าจะเข้าใจ — เปลี่ยนไปเขียน `for` loop ธรรมดาจะดีกว่า

### วนลูปผ่าน list และ tuple

วนตรง ๆ ได้เลย จะได้สมาชิกแต่ละตัวตามลำดับ (วิธีเดียวกันทั้ง list และ tuple)

### วนลูปผ่าน list/tuple พร้อม index ด้วย `enumerate()`

ถ้าต้องการทั้ง **index** และ **ค่า** ไปพร้อมกัน ใช้ `enumerate()` — ดีกว่าการใช้ `range(len(list))` แล้วเข้าถึงด้วย index มาก เพราะอ่านง่ายกว่า


````python
fruits = ["แอปเปิ้ล", "กล้วย", "ส้ม"]

# วิธีที่ไม่แนะนำ (อ่านยาก)
for i in range(len(fruits)):
    print(f"{i}: {fruits[i]}")

print()

# วิธีที่แนะนำ - ใช้ enumerate()
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")

print()

# enumerate() กำหนดเลขเริ่มต้นได้ด้วย (ปกติเริ่ม 0)
for i, fruit in enumerate(fruits, start=1):
    print(f"ลำดับที่ {i}: {fruit}")
````

### วนลูปผ่าน set

วนได้เหมือน list แต่**ลำดับที่ได้ไม่แน่นอน** (เพราะ set ไม่มีลำดับ) — ห้ามเขียนโค้ดที่พึ่งพาลำดับการวนของ set


````python
fruits_set = {"แอปเปิ้ล", "กล้วย", "ส้ม"}

for fruit in fruits_set:
    print(fruit)

# หมายเหตุ: ถ้ารันเซลล์นี้หลายรอบ ลำดับที่ print อาจไม่เหมือนกันทุกครั้ง
# (ขึ้นกับ Python version — Python รุ่นใหม่มักจะคงที่ในรอบการรันเดียวกัน แต่ไม่ควรเขียนโค้ดที่พึ่งพาลำดับนี้)
````

### วนลูปผ่าน dict — 3 รูปแบบที่ต้องรู้

มี 3 วิธีหลักในการวน dict ขึ้นอยู่กับว่าต้องการ key, value, หรือทั้งคู่


````python
student = {"name": "มานี", "age": 20, "gpa": 3.75}

# วิธี 1: วนผ่าน key เท่านั้น (ค่า default ถ้าไม่ระบุอะไร)
print("--- วนผ่าน key ---")
for key in student:
    print(key)

print()

# วิธี 2: วนผ่าน value เท่านั้น
print("--- วนผ่าน value ---")
for value in student.values():
    print(value)

print()

# วิธี 3: วนผ่านทั้ง key และ value พร้อมกัน (ใช้บ่อยที่สุด!)
print("--- วนผ่าน key และ value ---")
for key, value in student.items():
    print(f"{key}: {value}")
````

### วนลูปผ่าน nested dict (โครงสร้างซับซ้อนขึ้น)

ตัวอย่างที่ใช้บ่อยมากในการทำงานกับข้อมูลจริง (เช่นข้อมูลจาก API ที่มักเป็น nested dict)


````python
students = {
    "65010001": {"name": "มานี", "gpa": 3.75},
    "65010002": {"name": "สมชาย", "gpa": 3.20},
    "65010003": {"name": "วิชัย", "gpa": 2.95},
}

for student_id, info in students.items():
    print(f"รหัส {student_id}: {info['name']} (GPA {info['gpa']})")
````

### วนลูปผ่าน 2 collection พร้อมกันด้วย `zip()`

`zip()` ใช้ "รวม" หลาย list/tuple เข้าด้วยกันเป็นคู่ๆ ตามลำดับ — มีประโยชน์มากเวลามีข้อมูลหลาย list ที่สัมพันธ์กัน


````python
names = ["มานี", "สมชาย", "วิชัย"]
scores = [85, 90, 78]
grades = ["B+", "A", "C+"]

# zip() รวม 3 list เข้าด้วยกัน วนพร้อมกันทีละชุด
for name, score, grade in zip(names, scores, grades):
    print(f"{name}: คะแนน {score} เกรด {grade}")
````

**ข้อควรระวัง:** ถ้า list ที่ `zip()` มีความยาวไม่เท่ากัน — `zip()` จะหยุดตามตัวที่สั้นที่สุด (ไม่ error แต่อาจทำข้อมูลขาดไปโดยไม่รู้ตัว)


````python
short_list = [1, 2, 3]
long_list = ["a", "b", "c", "d", "e"]

for num, letter in zip(short_list, long_list):
    print(num, letter)
# สังเกตว่า "d" และ "e" จะไม่ถูกใช้เลย เพราะ short_list หมดก่อน
````

### สรุปวิธีวนลูปผ่านแต่ละ Collection

| Collection | วิธีวนพื้นฐาน | วิธีวนพร้อมข้อมูลเสริม |
|---|---|---|
| `list` / `tuple` | `for x in collection:` | `for i, x in enumerate(collection):` |
| `set` | `for x in collection:` (ลำดับไม่แน่นอน) | - |
| `dict` | `for key in collection:` | `for key, value in collection.items():` |
| **2 collections พร้อมกัน** | - | `for a, b in zip(list1, list2):` |

### ข้อ: ท้าทายเพิ่มเติม (Challenge) — FizzBuzz
เขียนโปรแกรมวนเลข 1 ถึง 30 โดยมีเงื่อนไข:
- ถ้าหารด้วย 3 ลงตัว → print "Fizz"
- ถ้าหารด้วย 5 ลงตัว → print "Buzz"
- ถ้าหารด้วยทั้ง 3 และ 5 ลงตัว → print "FizzBuzz"
- นอกจากนั้น → print ตัวเลขนั้นตามปกติ

(โจทย์นี้เป็นโจทย์สัมภาษณ์งานเขียนโปรแกรมที่คลาสสิกมาก ลองทำดู!)


````python
# เขียนคำตอบที่นี่

for i in range(1,30+1):

  if i % 3 == 0 and i % 5 == 0:
    print("FizzBuzz")
  elif i % 3 == 0:
    print("Fizz")
  elif i % 5 == 0:
    print("Buzz")
  else:
    print(i)
````
